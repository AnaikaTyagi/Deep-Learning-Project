```python
# =========================
# 0) Install deps (Colab)
# =========================
!pip -q install "transformers>=4.38" datasets peft trl accelerate bitsandbytes

import re
import random
import torch
from datasets import load_dataset, Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    BitsAndBytesConfig,
    TrainingArguments,
)
from peft import LoraConfig, get_peft_model
from trl import SFTTrainer


# =========================
# 1) Load + build dataset
# =========================
print("Loading TriviaQA...")
trivia = load_dataset("mandarjoshi/trivia_qa", "rc", split="train")

print("Loading SQuAD 2.0...")
squad = load_dataset("rajpurkar/squad_v2", split="train")

ABSTAIN_STR = "FINAL: I don't know."

def normalize(s: str) -> str:
    return re.sub(r"\s+", " ", s.strip().lower())

def format_trivia(example):
    # Ensure the answer is actually in the context (important!)
    question = example["question"]
    answer = example["answer"]["value"]
    ans_norm = normalize(answer)

    contexts = example["entity_pages"]["wiki_context"] or []
    chosen = None
    for ctx in contexts:
        if ans_norm and ans_norm in normalize(ctx):
            chosen = ctx
            break

    if not chosen:
        return None

    return {
        "prompt": f"Context: {chosen}\nQuestion: {question}\n",
        "response": f"FINAL: {answer}",
        "label": "answerable",
    }

def format_squad_unanswerable(example):
    # Unanswerable if there are no answer spans
    if len(example["answers"]["text"]) == 0:
        return {
            "prompt": f"Context: {example['context']}\nQuestion: {example['question']}\n",
            "response": ABSTAIN_STR,
            "label": "unanswerable",
        }
    return None

print("Formatting TriviaQA...")
trivia_formatted = []
for ex in trivia.select(range(20000)):
    out = format_trivia(ex)
    if out:
        trivia_formatted.append(out)

print("Formatting SQuAD unanswerables...")
squad_unanswerable = []
for ex in squad:
    out = format_squad_unanswerable(ex)
    if out:
        squad_unanswerable.append(out)
    if len(squad_unanswerable) >= 7000:
        break

full_dataset = trivia_formatted + squad_unanswerable
random.shuffle(full_dataset)

print(f"Total examples: {len(full_dataset)}")
print(f"Answerable: {sum(1 for x in full_dataset if x['label']=='answerable')}")
print(f"Unanswerable: {sum(1 for x in full_dataset if x['label']=='unanswerable')}")

# Convert into HF Dataset + make a single training text field
hf_ds = Dataset.from_list(full_dataset)

def to_text(example):
    # Single string for causal LM fine-tuning
    example["text"] = example["prompt"] + example["response"]
    return example

hf_ds = hf_ds.map(to_text, remove_columns=[])

# Train/val split
split = hf_ds.train_test_split(test_size=0.05, seed=42)
train_ds = split["train"]
eval_ds  = split["test"]


# =========================
# 2) Load model in 4-bit (QLoRA base)
# =========================
# Pick ONE model. Mistral-7B-Instruct-v0.3 is a common choice. (Requires HF access if gated in your environment.)
model_name = "mistralai/Mistral-7B-Instruct-v0.3"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.bfloat16 if torch.cuda.is_available() else torch.float16,
)

tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)
# Ensure a pad token exists for batching
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map="auto",
)

# =========================
# 3) Add LoRA adapters (this is the "trainable part")
# =========================
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM",
    # Common target modules for Mistral/LLaMA-style architectures:
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj", "gate_proj", "up_proj", "down_proj"],
)

model = get_peft_model(model, lora_config)
model.print_trainable_parameters()


# =========================
# 4) Train with SFTTrainer (supervised fine-tuning)
# =========================
max_seq_length = 512

training_args = TrainingArguments(
    output_dir="./abstain-qlora",
    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,
    gradient_accumulation_steps=8,   # helps on smaller GPUs
    learning_rate=2e-4,
    num_train_epochs=1,              # bump to 2-3 for better results
    lr_scheduler_type="cosine",
    warmup_ratio=0.03,
    logging_steps=25,
    eval_strategy="steps",
    eval_steps=200,
    save_steps=200,
    save_total_limit=2,
    bf16=torch.cuda.is_available(),  # if your GPU supports bf16
    fp16=not torch.cuda.is_available(),
    report_to="none",
)

trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=train_ds,
    eval_dataset=eval_ds,
    dataset_text_field="text",
    max_seq_length=max_seq_length,
    packing=False,
    args=training_args,
)

trainer.train()

# Save LoRA adapter weights (small!)
trainer.model.save_pretrained("./abstain-qlora-adapter")
tokenizer.save_pretrained("./abstain-qlora-adapter")


# =========================
# 5) Quick evaluation (your project metrics)
# =========================
@torch.no_grad()
def generate_final(prompt: str, max_new_tokens=32):
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    out = model.generate(
        **inputs,
        max_new_tokens=max_new_tokens,
        do_sample=False,   # greedy decoding like your proposal
        temperature=1.0,
        pad_token_id=tokenizer.eos_token_id,
    )
    text = tokenizer.decode(out[0], skip_special_tokens=True)
    # Return only the tail after the prompt
    return text[len(prompt):].strip()

def is_abstain(s: str) -> bool:
    return normalize(s).startswith(normalize(ABSTAIN_STR))

def extract_final_answer(s: str) -> str:
    m = re.search(r"FINAL:\s*(.*)", s, flags=re.IGNORECASE)
    return m.group(1).strip() if m else s.strip()

def exact_match(pred: str, gold: str) -> bool:
    return normalize(pred) == normalize(gold)

# Evaluate on a small slice to keep it fast
sample = eval_ds.select(range(min(400, len(eval_ds))))

tp_ans = 0     # correct answers on answerable
ans_total = 0
refuse_on_unans = 0
unans_total = 0
over_refuse = 0         # abstain on answerable
hallucinate = 0         # answer on unanswerable (proxy)

for ex in sample:
    prompt = ex["prompt"]
    gold_resp = ex["response"]
    label = ex["label"]

    pred_resp = generate_final(prompt)
    pred_final = extract_final_answer(pred_resp)
    gold_final = extract_final_answer(gold_resp)

    pred_is_abstain = is_abstain(pred_resp)

    if label == "answerable":
        ans_total += 1
        if pred_is_abstain:
            over_refuse += 1
        else:
            if exact_match(pred_final, gold_final):
                tp_ans += 1

    else:
        unans_total += 1
        if pred_is_abstain:
            refuse_on_unans += 1
        else:
            hallucinate += 1

acc_answerable = tp_ans / max(1, ans_total)
refusal_rate_unans = refuse_on_unans / max(1, unans_total)
over_refusal_rate = over_refuse / max(1, ans_total)
hallucination_rate = hallucinate / max(1, unans_total)

print("\n==== Metrics (quick eval) ====")
print(f"Accuracy (answerable):        {acc_answerable:.3f}")
print(f"Refusal rate (unanswerable):  {refusal_rate_unans:.3f}")
print(f"Over-refusal (answerable):    {over_refusal_rate:.3f}")
print(f"Hallucination proxy (unans):  {hallucination_rate:.3f}")

print("\n--- Example generations ---")
for i in range(3):
    ex = sample[i]
    pred = generate_final(ex["prompt"])
    print("\nLABEL:", ex["label"])
    print("PROMPT:", ex["prompt"][:250], "...")
    print("PRED:", pred)
    print("GOLD:", ex["response"])
```
